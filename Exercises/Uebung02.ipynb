## Aufgabe 1

a) Generieren Sie eine Liste von allen Download-URLs für 10cm SWISSIMAGE Bilder **im Kanton Bern**. (nur die Liste, kein Download!)<br/>
(für die Abfrage soll das generalisierte Polygon des Kanton Bern vom BFS verwendet werden)



```python
import geopandas as gpd
import shapely
import geopandas_stac as stac 
```

```python
df = stac.getAssets("ch.swisstopo.swissimage-dop10")
```

```python
df.head()
```

```python
df_10 = df[df['gsd']==0.1]
df_10
```

```python
kantone = gpd.read_file("daten/gemeindegrenzen/ggg_2021-LV95/shp/g1k21.shp", encoding="utf-8")
kantone = kantone[['KTNR','KTNAME','AREA_HA', 'geometry']]
kantone.head()
```

```python
bern = kantone[kantone['KTNAME']=='Bern / Berne']
bern
```

```python
# !!! Umwandlung !!!
bern = bern.to_crs("EPSG:4326")
bern.head()
```

```python
df_bern_10 = gpd.sjoin(df_10, bern, op='intersects')
```

```python
url_bern_10 = stac.getUrlList(df_bern_10)
url_bern_10[0:5]
```

b) Wieviele Bilder sind das ?


```python
len(url_bern_10)
```


c) Wieviel Disk Speicher würde bei einem Download aller Files ungefähr benötigt ?


```python
import urllib.request

file = urllib.request.urlopen(url_bern_10[0])
```

```python
url_bern_10[0]
```

```python
s = file.length
print(s, "B")
print(s/1024, "KB")
print(s/1024/1024, "MB")
print("Alle Files sind ungefähr",s/1024/1024/1024*len(url_bern_10), "GB gross")
```

## Aufgabe 2

Bestimmen Sie die URL der swissALTI3D 0.5cm an einer beliebigen Position (WGS84) in der Schweiz. (nur URL, kein Download)

```python
df = stac.getAssets("ch.swisstopo.swissalti3d")
```

```python
df.head()
```

```python
print ( list(df['variant'].unique()) )
print ( list(df['type'].unique()) )
print ( list(df['gsd'].unique()) )
```

```python
df_05 = df[(df['gsd']==0.5) & (df['type']=='image/tiff; application=geotiff; profile=cloud-optimized')]
df_05.head()
```

```python
df_05.crs
```

```python
location = [47.35029409613778, 7.899542370040819]
```

```python
location[0]
```

```python
point = shapely.geometry.Point(location[1], location[0])
olten_punkt = gpd.GeoDataFrame(geometry=gpd.GeoSeries(point, crs="epsg:4326"))
```

```python
olten_punkt
```

```python
df_05_olten = gpd.sjoin(df_05, olten_punkt, op='contains')
```

```python
urls = stac.getUrlList(df_05_olten)
urls
```
