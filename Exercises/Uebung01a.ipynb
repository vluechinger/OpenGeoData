```python
import os

if not os.path.exists("daten"):
    os.mkdir("daten")
    
import folium

import urllib
import shutil
```

```python
file = open("daten/SBB_Bauprojekte.csv", encoding="utf-8")

i = 0

for line in file:
    print(line.strip())
    
    i+=1
    if i==3:
        break
    
file.close()
```

```python
m = folium.Map(location=[47.37825,8.5367835], tiles='', zoom_start=7)
 
folium.raster_layers.TileLayer(
    tiles="https://wmts.geo.admin.ch/1.0.0/ch.swisstopo.pixelkarte-farbe/default/current/3857/{z}/{x}/{y}.jpeg",
    attr="(c) swisstopo",
    name="pixelkarte-farbe",
    min_zoom=8,
    max_zoom=18,
    tms=False,
    overlay=False,
    control=False,
).add_to(m)

folium.LayerControl().add_to(m)


file = open("daten/SBB_Bauprojekte.csv", encoding="utf-8")

#geo_point_2d;projektname;projektort;angebotsschritt;link1_title;link1_url;link2_title;link3_url;link3_title;ort;art

next(file)
for line in file:
    data = line.strip().split(";")
    projektname = data[1]
    projektort = data[2]
    lnglat = data[0].split(",")  
    lat = float(lnglat[0])
    lng = float(lnglat[1])
    
    #m.get_root().html.add_child(folium.Element(f"{projektname}<br/>{projektort}"))

    folium.Marker([lat,lng], 
         popup=f"{projektname}<br/>{projektort}",
         icon=folium.Icon(color="red", prefix="fa", icon="star")).add_to(m)
    
file.close()


m.save("daten/SBB.html")
m
```

```python
from folium.plugins import MarkerCluster

m = folium.Map(location=[47.37825,8.5367835], zoom_start=7)

marker_cluster = MarkerCluster(control=True).add_to(m)
folium.LayerControl().add_to(m)

file = open("daten/SBB_Bauprojekte.csv", encoding="utf-8")

next(file)
for line in file:
    data = line.strip().split(";")
    projektname = data[1]
    projektort = data[2]
    lnglat = data[0].split(",")  
    lat = float(lnglat[0])
    lng = float(lnglat[1])
    folium.Marker([lat,lng], popup=f"{projektname}<br/>{projektort}", icon=folium.Icon(color="orange", prefix="fa", icon="star")).add_to(marker_cluster)
    
file.close()

m.save("daten/SBB_markercluster.html")
m

```
